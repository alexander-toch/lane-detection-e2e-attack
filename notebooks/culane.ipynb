```python
import numpy as np
np.set_printoptions(threshold=np.inf)
import cv2
import matplotlib.pyplot as plt

image = np.array(cv2.cvtColor(cv2.imread("00630.png"), cv2.COLOR_RGB2GRAY))
image = cv2.resize(image, (800, 288))

print(image.shape)

fig, ax = plt.subplots(figsize=(9, 3))
ax.imshow(image, 'gray', vmin=0, vmax=4)
ax.set_aspect('auto')
```
