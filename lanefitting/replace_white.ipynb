```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

image = np.array(cv2.cvtColor(cv2.imread("white_box.png"), cv2.COLOR_BGR2RGB))
patch = np.array(cv2.cvtColor(cv2.imread("patch_100.jpg"), cv2.COLOR_BGR2RGB))

# find white pixels
white = np.where(np.all(image == [255, 255, 255], axis=-1))

# find bounding box
min_x = np.min(white[1])
max_x = np.max(white[1])
min_y = np.min(white[0])
max_y = np.max(white[0])


# resize patch to fit bounding box
patch = cv2.resize(patch, (max_x - min_x, max_y - min_y))

# fill area with patch
image[min_y:max_y, min_x:max_x] = patch

plt.imshow(image)
```
